```python
# Importing the Required Libraries
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
pd.set_option("display.max_columns", None)
```

```python
# Loading the Initial Stocks Database
df_train = pd.read_csv('/Users/shashwatmishra/Desktop/Stocks_final.csv')
df_train
```

```python
# Showing plots of the features, since there are 300 features, showing only 25 to get an overview
plt.figure(figsize=(20,20))
for i in range(25) :
    plt.subplot(5, 5, i+1)
    plt.hist(df_train[f"f_{i}"], bins=100)
    plt.title(f"f_{i}")
plt.show
```

```python
# Describing the Data
df_train.describe()
```

```python
feature_cols = df_train.columns.unique() # initializing columns containing only features
feature_cols = feature_cols.drop(["row_id", "time_id", "target"]) # dropping all columns other than features
feature_cols
```

```python
# Dtecting the outliers from the features using std.
list_of_outliers = []
col_of_outliers = []

for cols in feature_cols:
    temp_df = df_train[(df_train[cols] > df_train[cols].mean() + df_train[cols].std() * 40) |
                       (df_train[cols] < df_train[cols].mean() - df_train[cols].std() * 40) ]
    
    temp2_df = df_train[(df_train[cols] > df_train[cols].mean() + df_train[cols].std() * 20) |
                       (df_train[cols] < df_train[cols].mean() - df_train[cols].std() * 20) ]
    if len(temp_df) >0 :
        outliers = temp_df.index.to_list()
        list_of_outliers.extend(outliers)
        col_of_outliers.append(cols)
        print(cols, len(temp_df))
    
    elif len(temp2_df)>0 and len(temp2_df) <6 :
        outliers = temp2_df.index.to_list()
        list_of_outliers.extend(outliers)
        col_of_outliers.append(cols)
        print(cols, len(temp2_df))

list_of_outliers = list(set(list_of_outliers))
print(len(col_of_outliers),len(list_of_outliers))
```

```python
# Plotting the feature before removing the detected outliers
plt.figure(figsize=(20,20))
for i, (col) in enumerate(["f_87", "f_117", "f_289"]) :
    plt.subplot(5, 5, i+1)
    plt.scatter(df_train[col], df_train["target"])
    plt.title(col)
plt.show
```

```python
# Removing the Outliers Detected
df_train.drop(df_train.index[list_of_outliers], inplace = True)
df_train
```

```python
# Plotting Example Features after Outliers are Removed
plt.figure(figsize=(20,20))
for i, (col) in enumerate(["f_87", "f_117", "f_289"]) :
    plt.subplot(5, 5, i+1)
    plt.scatter(df_train[col], df_train["target"])
    plt.title(col)
plt.show
```

```python
# Custom Feature/Attribute Selection

X = df_train.loc[:, df_train.columns != 'row_id']
X = X.loc[:, X.columns != 'time_id']
X = X.loc[:, X.columns != 'investment_id']

# Normalization
X=(X-X.mean())/X.std() 

# Setting Predictor and Target Attribute
y = X["target"]
X = X.loc[:, X.columns != 'target']
print(X)
print(y)
```

```python
from sklearn import linear_model
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import  mean_absolute_error,mean_squared_error 

# Setting Multiple Prediction Model Into loops to use to predict the target value
classifiers = [
    svm.SVR(),
    linear_model.LinearRegression(),
    linear_model.BayesianRidge(),
    linear_model.PassiveAggressiveRegressor()]

# Splitting the Data For Training and Tesing with the ratio of 80% to 20%
X_train, X_test, y_train, y_test =  train_test_split(X, y, train_size=0.8, test_size=0.2)

for item in classifiers:
    print(item)
    clf = item
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    
    # Displaying the Predicted Target Values for each of the Model
    print(clf.predict(X_test),'\n')
    
    # Displaying the Mean Absolute Error for Each of the Prediction Model
    print('mean_absolute_error: \n')
    print(mean_absolute_error(y_test, y_pred),'\n')
    
    # Displaying the Predicted to Absolute Value Plot
    plt.figure(figsize=(10,10))
    plt.scatter(y_test, y_pred, c='crimson')
    plt.yscale('log')
    plt.xscale('log')

    p1 = max(max(y_pred), max(y_test))
    p2 = min(min(y_pred), min(y_test))
    plt.plot([p1, p2], [p1, p2], 'b-')
    plt.xlabel('True Values', fontsize=15)
    plt.ylabel('Predictions', fontsize=15)
    plt.axis('equal')
    plt.show()
```
